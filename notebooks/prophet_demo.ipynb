```python
import pandas as pd
from prophet import Prophet
import random
```

```python
y = [random.randint(0, 100) for i in range(100000)]
df = pd.DataFrame(data={'y': y},)
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365, include_history=False)
```

```python
future
```

```python
forecast = m.predict(future)
```

```python
forecast
```

```python
m.plot(forecast)
```
